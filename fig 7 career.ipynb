```python
f1['citing_year'] = f1['citing_id'].map(f2['year'])
```

```python
def calculateDI(focal_paper , year):
    f11 = f1[f1['citing_year'] <= year].copy()
    citing_lst = f11[f11['cited_id']==focal_paper][f11['citing_year'] == year]['citing_id'].values
    ref_lst = f11[f11['citing_id']==focal_paper]['cited_id'].values
    if len(ref_lst)>0:
        ref_citing_lst = np.concatenate([ f11[f11['cited_id'] == i][f11['citing_year'] == year]['citing_id'].values for i in ref_lst])
        ref_citing_lst = np.unique(ref_citing_lst)
    else:
        ref_citing_lst = np.array([])
    
    nj = np.intersect1d(citing_lst,ref_citing_lst)
    ni = np.setdiff1d(citing_lst,nj)
    
    return len(ni) , len(nj)
```

```python
dic = dict(zip(f2.index , f2['year']))
func = lambda x:dic[x]

def get_year(pubarray):
    pubyear = np.vectorize(func)(pubarray)
    pubyear.sort()
    return pubyear
```

```python
def get_DCCClst(i):
    pubarray = flau['pub'].values[i]
    pubyear = get_year(pubarray)
    DClst = []
    CClst = []
    DC = 0
    CC = 0
    for year in range(pubyear[0] , pubyear[0]+40):
        
        for j in range(len(pubarray)):
            if pubyear[j]<=year:
                DCj , CCj = calculateDI(pubarray[j] , year)
                DC += DCj
                CC += CCj
        DClst.append(DC)
        CClst.append(CC)
    return [DClst , CClst]
```

```python
DCCClst = [get_DCCClst(i) for i in range(len(flau))]
```

```python
flau['DCCClst'] = DCCClst
```

```python
flau['DClst'] = flau['DCCClst'].apply(lambda x:x[0])
flau['CClst'] = flau['DCCClst'].apply(lambda x:x[1])
```
