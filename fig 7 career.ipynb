```python
flau = fau[fau['type']=='laureate'].copy()
```

```python
def evalh(x):
    if type(x)==str:
        return eval(x)
    return []
```

```python
flau['pub'] = flau['pub'].apply(evalh)
flau['DClst'] = flau['DClst'].apply(evalh)
flau['CClst'] = flau['CClst'].apply(evalh)
```

```python
flau['au'] = flau['au'].apply(lambda x:x.upper())
```

```python
data1 = np.array([np.array(i) for i in flau['DClst'].values])
data2 = np.array([np.array(i) for i in flau['CClst'].values])
```

```python
# Create x-axis values
x = np.arange(1, 41)
# Calculate the mean and standard deviation for each column
mean1 = np.mean(data1, axis=0)
std1 = np.std(data1, axis=0)
ci1 = scipy.scipy.stats.bootstrap
# Create y-axis values (means)
y1 = mean1

# Calculate the mean and standard deviation for each column
mean2 = np.mean(data2, axis=0)
std2 = np.std(data2, axis=0)
# Create y-axis values (means)
y2 = mean2
```

```python
f, ax = plt.subplots(figsize=(14, 9))
# Plot the line graph with error bars
plt.plot(x, y1, color='#1A5276', linewidth=2)
plt.scatter(x, y1, s=200,marker='o',c='#1A5276',label = 'Disruptive Citation')
plt.errorbar(x, y1, yerr=std1, fmt='-', ecolor='#1A5276', capsize=5,alpha = 0.6)

plt.plot(x, y2, color='#AF601A', linewidth=2)
plt.scatter(x, y2, s=200,marker='v',c='#AF601A',label = 'Consolidating Citation')
plt.errorbar(x, y2, yerr=std2, fmt='-', ecolor='#AF601A', capsize=5,alpha = 0.6)
ax.set_yscale("log")
# Set the title and labels
plt.legend(fontsize = 25,loc = 'upper left')
plt.yticks(fontsize=20,)
plt.xticks(fontsize=20,)
plt.xlabel('Career Year',fontsize = 30)
plt.ylabel('Mean Value',fontsize = 30)
```
